1. What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly explain with examples.
- Paste

  Pastes everything: values, formulas, formatting.

  Example: Copy =A1+B1 → Paste → formula + formatting both come.

- Paste Special

  Lets you choose what to paste:

  Values only

  Formulas only

  Formats only

- Example: Copy calculated total → Paste Special → Values → only the number appears, not the formula.

2.  Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes' in Excel.
- Freeze Panes

  Locks rows/columns so headers stay visible while scrolling.

  Example: Freeze top row to always see column headings.

- Split Panes

  Divides worksheet into multiple scrollable sections.

  Example: View top and bottom of a long sheet at the same time.

- Use case difference:

  Freeze = keep headings visible

  Split = compare different parts of the sheet
   
3. Explain the difference between inserting a new row and inserting a new column in Excel. Can you insert multiple rows or columns at once?
- Insert Row: Adds a new horizontal row

  Insert Column: Adds a new vertical column

- Yes, we can insert multiple at once!

  Select 3 rows → Right-click → Insert → 3 new rows appear

  Same for columns

4. What are logical functions in Excel? Provide examples of at least two logical functions and their applications.
- Logical functions help Excel make decisions (TRUE/FALSE).

- Examples:

  =IF(A1>50,"Pass","Fail")
  → Used for grading students

  =AND(A1>50, B1>50)
  → Checks if both conditions are true

- Applications:

  Attendance status

  Pass/fail systems

  Bonus eligibility

5. Discuss the purpose of 'XLOOKUP' and how it differs from the traditional 'VLOOKUP' function.
- XLOOKUP

  Modern lookup function

  Can search left or right

  More flexible and accurate

- VLOOKUP

  Older function

  Can only search from left to right

  Breaks if columns move
